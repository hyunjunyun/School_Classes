> 1. 서식 지정자로 문자열 넣기
* %s
* %c
* %d
* %f


```
print("나는 %d살 입니다." % 10)
```

> 2. format 메서드로 문자열 넣기
* "{index}".format(값)



```
print('나는 {}살 {}입니다.' .format(20,'남자'))
print('나는 {1}살 {0}입니다.' .format(20,'남자'))
print('"{0:^10}"'.format('hi'))

age = 20
color = '빨간'

print("나는{}살이고 {}색을 좋아합니다." .format(age,color))
```
